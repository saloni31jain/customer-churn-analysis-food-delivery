```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/online_retail_II.csv", encoding="latin1")

df.head()
```

```python
df.info()
```

```python
df = df[[
    "Customer ID",
    "Invoice",
    "InvoiceDate",
    "Quantity",
    "Price"
]]
```

```python
df.columns = [
    "customer_id",
    "order_id",
    "order_date",
    "quantity",
    "price"
]
```

```python
df = df.dropna(subset=["customer_id"])
```

```python
df = df[df["quantity"] > 0]
```

```python
df["order_value"] = df["quantity"] * df["price"]
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], format="%m/%d/%Y %H:%M")
```

```python
df.head()
df.describe()
```

```python
df.to_csv("../data/clean_orders.csv", index=False)
```
